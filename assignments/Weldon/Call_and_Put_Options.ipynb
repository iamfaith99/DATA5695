```python
import numpy as np
import matplotlib.pyplot as plt
```

## Long Call Option

```python
# Prices, Payoffs, Strike, Premium
strike = 1000.0
call_premium = 93.81 * np.exp(0.02)
spot_prices = np.arange(800.0, 1250.0, 50.0)
call_payoffs = np.maximum(spot_prices - strike, 0.0)
call_profit = call_payoffs - call_premium
forward_price = 1020.0
long_payoffs = spot_prices - forward_price

# 
plt.figure(figsize=(10, 6))
plt.plot(spot_prices, call_profit, label='Long Call Profit', linestyle='--', linewidth=2)
plt.plot(spot_prices, long_payoffs, label='Long Forward Payoff', linestyle='--', linewidth=2)

# Adding labels, title, and legend
plt.axhline(0, color='black', linewidth=1, linestyle='--')  # Add horizontal line at payoff = 0
plt.xlabel('Value of Index at Expiration', fontsize=12)
plt.ylabel('Payoff', fontsize=12)
plt.title('(Long) Call Profit vs Long Forward Payoff', fontsize=14)
plt.legend(fontsize=10)
plt.grid(alpha=0.5)

# Show the plot
plt.show()
```

## Short Call Option

```python
# Prices, Payoffs, Strike, Premium
strike = 1000.0
call_premium = 93.81 * np.exp(0.02)
spot_prices = np.arange(800.0, 1250.0, 50.0)
# For short call, we negate the payoff and profit
short_call_payoffs = -np.maximum(spot_prices - strike, 0.0)
short_call_profit = short_call_payoffs + call_premium  # Note: we add premium since we receive it
forward_price = 1020.0
short_forward_payoffs = -(spot_prices - forward_price)  # Negative of long forward

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(spot_prices, short_call_profit, label='Short Call Profit', linestyle='--', linewidth=2)
plt.plot(spot_prices, short_forward_payoffs, label='Short Forward Payoff', linestyle='--', linewidth=2)

# Adding labels, title, and legend
plt.axhline(0, color='black', linewidth=1, linestyle='--')  # Add horizontal line at payoff = 0
plt.xlabel('Value of Index at Expiration', fontsize=12)
plt.ylabel('Payoff', fontsize=12)
plt.title('(Short) Call Profit vs Short Forward Payoff', fontsize=14)
plt.legend(fontsize=10)
plt.grid(alpha=0.5)
```

## Long Put Option

```python
# Long Put Option Plot
# Prices, Payoffs, Strike, Premium
strike = 1000.0
put_premium = 73.81 * np.exp(0.02)  # Using a sample put premium
spot_prices = np.arange(800.0, 1250.0, 50.0)
put_payoffs = np.maximum(strike - spot_prices, 0.0)  # Put payoff formula
put_profit = put_payoffs - put_premium  # Long put profit
forward_price = 1020.0
long_payoffs = spot_prices - forward_price

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(spot_prices, put_profit, label='Long Put Profit', linestyle='--', linewidth=2)
plt.plot(spot_prices, long_payoffs, label='Long Forward Payoff', linestyle='--', linewidth=2)

# Adding labels, title, and legend
plt.axhline(0, color='black', linewidth=1, linestyle='--')
plt.xlabel('Value of Index at Expiration', fontsize=12)
plt.ylabel('Payoff', fontsize=12)
plt.title('(Long) Put Profit vs Long Forward Payoff', fontsize=14)
plt.legend(fontsize=10)
plt.grid(alpha=0.5)
plt.show()
```

## Short Put Option

```python
# Short Put Option Plot
# Prices, Payoffs, Strike, Premium
strike = 1000.0
put_premium = 73.81 * np.exp(0.02)  # Using a sample put premium
spot_prices = np.arange(800.0, 1250.0, 50.0)
short_put_payoffs = -np.maximum(strike - spot_prices, 0.0)  # Negative of long put payoff
short_put_profit = short_put_payoffs + put_premium  # Add premium (received)
forward_price = 1020.0
short_forward_payoffs = -(spot_prices - forward_price)  # Negative of long forward

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(spot_prices, short_put_profit, label='Short Put Profit', linestyle='--', linewidth=2)
plt.plot(spot_prices, short_forward_payoffs, label='Short Forward Payoff', linestyle='--', linewidth=2)

# Adding labels, title, and legend
plt.axhline(0, color='black', linewidth=1, linestyle='--')
plt.xlabel('Value of Index at Expiration', fontsize=12)
plt.ylabel('Payoff', fontsize=12)
plt.title('(Short) Put Profit vs Short Forward Payoff', fontsize=14)
plt.legend(fontsize=10)
plt.grid(alpha=0.5)
plt.show()
```

```python
# Define the strike price
K = 50

# Generate a range of stock prices around the strike price
S = np.linspace(30, 70, 200)

# Calculate call option payoff: max(S - K, 0)
call_payoff = np.maximum(S - K, 0)

# Calculate ReLU applied to (S - K)
relu_values = np.maximum(S - K, 0)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the call option payoff
plt.plot(S, call_payoff, label=f'Call Option Payoff (K = {K})', color='blue', linewidth=2)

# Plot the ReLU function applied to (S - K)
plt.plot(S, relu_values, '--', label='ReLU(S - K)', color='red', linewidth=2)

# Add labels and title
plt.xlabel('Stock Price at Expiration (S)', fontsize=12)
plt.ylabel('Payoff / ReLU Value', fontsize=12)
plt.title('Relationship Between Call Option Payoff and ReLU Function', fontsize=14)

# Add a vertical line at the strike price
plt.axvline(x=K, color='green', linestyle=':', linewidth=1.5, label=f'Strike Price (K = {K})')

# Add grid and legend
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=10)

# Show the plot
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
relu = np.maximum(0, x)

plt.figure(figsize=(8, 4))
plt.plot(x, relu, label='ReLU(x)', color='blue', linewidth=2)
plt.title('ReLU Activation Function', fontsize=14)
plt.xlabel('Input (x)', fontsize=12)
plt.ylabel('Output', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
K = 50  # Strike price
S = np.linspace(30, 70, 200)  # Stock prices

# Define payoffs
long_call = np.maximum(S - K, 0)
short_call = -long_call
long_put = np.maximum(K - S, 0)
short_put = -long_put

# Plot all payoffs
plt.figure(figsize=(12, 7))

plt.plot(S, long_call, label='Long Call', color='blue', linewidth=2)
plt.plot(S, short_call, '--', label='Short Call', color='red', linewidth=2)
plt.plot(S, long_put, label='Long Put', color='green', linewidth=2)
plt.plot(S, short_put, '--', label='Short Put', color='purple', linewidth=2)

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(K, color='gray', linestyle=':', label=f'Strike Price (K={K})')

plt.xlabel('Stock Price (S)', fontsize=12)
plt.ylabel('Payoff', fontsize=12)
plt.title('Option Payoffs vs. ReLU Variations', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
